```python
### create pooling function; output is average of all words in review
## 
#@jit(nopython=True,parallel=True) // numba somehow not working due to incompatibility of version... 
def create_review_embedding(dataO,enc):
    outvec = np.zeros(400)
    for i in dataO:
         outvec = np.vstack((outvec,enc[i].mean(axis=0)))
    return outvec
        
    
    

```

```python
vecs = create_review_embedding(dataO,enc.values)
```

```python
vecs = pd.DataFrame(vecs)
vecs.to_pickle("trainng_set_encoded")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
n_reviews =None
ixx = vecs.index
```

```python
def highlight_most_representative_reviews(vecs,n_reviews=None,ixx=vecs.index):
    vecS = vecs.iloc[ixx,:].copy()
    if n_reviews is None:
        n_reviews = np.int(vecS.shape[0]**0.5)
    km = KMeans(n_clusters=n_reviews)
    km = km.fit(vecS)
    closeness,_ = pairwise_distances_argmin_min(km.cluster_centers_,vecS)
    return closeness
    
    
```

```python

ix = highlight_most_representative_reviews(vecs,n_reviews=15)
```
